```python
import matplotlib.pyplot as plt
%matplotlib inline

class Circle(object):
    def __init__(self, radius, color):
        self.radius=radius
        self.color=color
    def add_radius(self,r):
        self.radius=self.radius+r
        return(self.radius)
    def draw_circle(self):
        #object attached to a plot
        drawC=plt.Circle((0,0),radius=self.radius,fc=self.color)
        # access a plot area or the last created plot
        ax= plt.gca()
        # a patch refers to colored area--> for any shapes except lines
        ax.add_patch(drawC)
        ax.axis("scaled")
        plt.grid(color="grey", linestyle="--")
        plt.show()
class Rectangle(object):
    def __init__(self,height,width,color):
        self.height=height
        self.width=width
        self.color=color
    def draw_rectangle(self):
        draw_Rect=plt.Rectangle((0,0),height=self.height,width=self.width,fc=self.color)
        ax=plt.gca()
        ax.add_patch(draw_Rect)
        ax.axis("scaled")
        plt.grid("True")
        plt.grid(color="grey", linestyle="--")
        plt.show()        
Red_circle= Circle(4,'red')
print (Red_circle.radius)
print(Red_circle.color)
Red_circle.draw_circle()
rad=Red_circle.add_radius(3)
print(rad)
Red_circle.draw_circle()
green_rectangle= Rectangle(4,5,"yellow")
green_rectangle.draw_rectangle()
```
